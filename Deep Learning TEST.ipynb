```python
## Divide the training dataset into train and validation data along with normalization

X_valid,X_train = X_train_full[:5000]/255 , X_train_full[5000:]/255
y_valid,y_train = y_train_full[:5000]/255 , y_train_full[5000:]/255

X_test = X_test/255
```

```python
## Creating Layers of ANN
LAYERS = [
    tf.keras.layers.Flatten(input_shape=[28,28],name='input_layer'),
    tf.keras.layers.Dense(300,activation='relu',name='hidder_layer1'),
    tf.keras.layers.Dense(100,activation='relu',name='hidder_layer2'),
    tf.keras.layers.Dense(10,activation='softmax',name='output_layer')]

mod_clf= tf.keras.models.Sequential(LAYERS)
```

```python
LOSS_FUNCTION = tf.losses.SparseCategoricalCrossentropy()
OPTIMIZER = 'SGD'
METRICES = 'accuracy'
```

```python
mod_clf.compile(loss=LOSS_FUNCTION,optimizer=OPTIMIZER,metrics=METRICES)
```

```python
# Training

EPOCHS = 30
VALIDATION_SET = (X_valid,y_valid)
history = mod_clf.fit(X_train,y_train,batch_size=32,validation_data=VALIDATION_SET,epochs=EPOCHS)
```
